```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('C:\Practice_code\Gfg\housingprice 2.csv' ,delimiter=",")
x = data[:,0]#area of the house
y = data[:,1]#price of the house
m = len(y)

#plot of data
plt.figure(figsize=(8,6))
plt.plot(x,y,'rx',markersize=5)
plt.xlabel('Area of house')
plt.ylabel('Price of  house')
plt.title('Scatter plot of data')
plt.show()

#prepare the data
x = np.stack([np.ones(m),x],axis =1)#add a column of ones to X
print(X)

#compute the error/cost function
def compute_cost(x,y,theta):
    predictions =x @ theta
    error = predictions-y
    cost = (1/(2*m))*np.dot(error,error)
    return cost

#gradient descent
def gradient_descent(x,y,theta,alpha,iterations):
    cost_history =[]

    for i in range(iterations):
        predictions = x @ theta
        error = predictions-y
        gradient = (1/m)*(x.T@ error)
        theta = theta -alpha*gradient
        cost = compute_cost(x,y,theta)
        cost_history.append(cost)

    return theta,cost_history

theta =[0,0]
iterations = 150
alpha = 0.00001
theta,cost_history = gradient_descent(x,y,theta,alpha,iterations)
print('theta found by GDA',theta)

plt.figure(figsize=(8,6))
plt.plot(x[:,1],y,'rx',markersize = 5)
plt.plot(x[:,1],x@theta,'_')
plt.xlabel('Area of house')
plt.ylabel('Housing price')
plt.legend()
plt.title("scatter plot ")
plt.show()
```
